```python
from statistics import mean
#pd.plotting.register_matplotlib_converters()

df_dates = df.dropna(subset=['fecha'])

df_dates['year_month'] = df_dates['fecha'].dt.to_period('M')
month_count = df_dates['year_month'].value_counts().sort_index()

count = month_count.to_list()
month = month_count.keys().to_series().astype(str)
mean_score = mean(count)

# plot
fig, ax = plt.subplots(figsize=(14, 8))
bars0 = ax.bar(month, count, color='maroon', alpha=0.6, edgecolor=colors_dark[0])
#plt.plot(month,count,linestyle='-', marker='o')
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Fecha", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel('Número de siniestros', fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
plt.xticks(rotation='vertical')
line  = ax.axhline(mean_score, linestyle='--', color=colors_dark[2])
avgl  = ax.text(
    s="Promedio\nGlobal: {:.2f}".format(mean_score),
    x=xmax*1.02,
    y=mean_score,
    backgroundcolor=colors_dark[2],
    fontsize=12,
    fontweight='bold',
    color='white'
)


plt.text(s="¿En qué meses ocurren más accidentes?", ha='left', x=xmin, y=ymax*1.12, fontsize=24, fontweight='bold', color=colors_dark[0])       
plt.title("Distribución mensual de los siniestros viales ocurridos entre {:%d/%m/%Y} y {:%d/%m/%Y} en CABA".format(df['fecha'].min(),df['fecha'].max()), loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```

```python
from datetime import datetime as dt
df_time = df.dropna(subset=['hora'])
span = (df['fecha'].max() - df['fecha'].min()).days
df_hours = pd.to_datetime(df_time['hora'],format='%H:%M:%S').dt.round("H").dt.time.apply(lambda x: x.strftime('%H:%M:%S'))
hours = (df_hours.value_counts()/span).sort_index()

count = hours.to_list()
hour = hours.keys().to_list()
mean_score = hours.mean()

fig, ax = plt.subplots(figsize=(14, 8))
# plot
plt.plot(hour,count,linestyle='-', marker='o')

xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Horario", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel('Número de siniestros por hora', fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
line  = ax.axhline(mean_score, linestyle='--', color=colors_dark[2])
avgl  = ax.text(
    s="Promedio\nGlobal: {:.2f}".format(mean_score),
    x=xmax*1.02,
    y=mean_score,
    backgroundcolor=colors_dark[2],
    fontsize=12,
    fontweight='bold',
    color='white'
)

# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.text(s="¿En qué horarios ocurren más accidentes?", ha='left', x=xmin, y=ymax*1.12, fontsize=24, fontweight='bold', color=colors_dark[0])       
plt.title("Distribución por hora de los siniestros viales ocurridos entre {:%d/%m/%Y} y {:%d/%m/%Y} en CABA".format(df['fecha'].min(),df['fecha'].max()), loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```

```python
from shapely.geometry import Point
import geopandas as gpd

df_coor = df.dropna(subset=['x','y'])
comunas = gpd.read_file(r"C:\Users\ivans\Documents\Data Analysis\Accidentes\comunas.csv")

def coordenate_classifier(x, y, multipols):
    density = [0] * len(multipols)
    xy = list(zip(df_coor.x,df_coor.y))
    for coor in xy:
        point = Point(coor)
        for i, geom in enumerate(multipols):
            if geom.contains(point):
                density[i] += 1
    return density

den = coordenate_classifier(df_coor.x, df_coor.y, comunas.geometry)
```

```python
import matplotlib.colors as mcolors

comunas['siniestros'] = den

fig, ax = plt.subplots(figsize=(14, 8))
plt.xticks(rotation='vertical', fontsize=10)
plt.yticks(fontsize=10)
ax.grid(axis='both', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Latitud [º]", fontsize=11, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Longitud [º]", fontsize=11, labelpad=10, fontweight='bold', color=colors_dark[0])

comunas.plot(column='siniestros', ax=ax, legend=True, vmin=0, vmax = 4500, cmap='Purples', legend_kwds={'label': "Número de siniestros",'orientation': "vertical"})

xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

plt.text(s="¿En qué barrios ocurren más accidentes?", ha='left', x=xmin, y=ymax/1.0005, fontsize=16, fontweight='bold', color=colors_dark[0])       
plt.title("Distribución por barrio de los siniestros viales ocurridos entre {:%d/%m/%Y} y {:%d/%m/%Y} en CABA".format(df['fecha'].min(),df['fecha'].max()), loc='left', fontsize=9, color=colors_dark[2])  
plt.tight_layout()
plt.show()
plt.show()
```
